## Function

```python
def checkComponentWSSE(data):
    scale = RobustScaler()
    data_scale = scale.fit_transform(data)
    length_columns = data.shape[1]
    dict_df = {}
    for col in range(0, length_columns):
        dict_df[col] = data_scale[: , col]
    df = pd.DataFrame(dict_df)
    for number in range(0, length_columns):
        columns = list(df.columns)
        col = columns.pop(number)
        df_sub = df[columns]
        showTestCluster(df_sub, 10, 'Remove: ' + data.columns[col])
```

```python
def buildModel(data, k):
    preprocess = make_column_transformer(
        (RobustScaler(), data.columns),
    )
    pipeline = make_pipeline(
        preprocess,
        KMeans(n_clusters=k, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    )
    return pipeline.fit(data)
```

```python
def showCorr(title, data_original):
    tuple_name = set(
        data_original.Name  
    )
    columns = [
        'Ten Tinh', 
        'So_Diem_Chay VS Sunshine', 
        'So_Diem_Chay VS Air_Humidity', 
        'So_Diem_Chay VS Air_Temperature', 
        'So_Diem_Chay VS Rainfull', 
    ]
    df_corr = pd.DataFrame(columns=columns)
    for name in tuple_name:
        df = data_original[data_original.Name == name]
        new_row = pd.Series(
            data = {
                columns[0]: name,
                columns[1]: df.So_Diem_Chay.corr(df.Sunshine),
                columns[2]: df.So_Diem_Chay.corr(df.Air_Humidity),
                columns[3]: df.So_Diem_Chay.corr(df.Air_Temperature),
                columns[4]: df.So_Diem_Chay.corr(df.Rainfull)
            }
        )
        df_corr = df_corr.append(new_row, ignore_index=True)
    df_corr_show = df_corr.set_index('Ten Tinh')
    plt.figure(figsize=(10, 2))
    sb.heatmap(
        df_corr_show.T,
        annot=True,
        vmax=1,
        vmin=-1,
        fmt='0.1f',
        cmap='coolwarm'
    )
    plt.title(title + ': Mối quan hệ giữa các chỉ số thời tiết với số Điểm cháy qua các tỉnh', fontweight='bold', fontsize=14)
    plt.show()
```

```python
def showPCA(df):
    pca = PCA(n_components=2)
    df_pca = pca.fit_transform(df[[
        'So_Diem_Chay', 'Sunshine', 'Air_Humidity', 'Air_Temperature', 'Rainfull'
    ]])
    df_pca = pd.DataFrame(data = df_pca, columns = ['1', '2'])
    # -----------
    plt.figure(figsize=(5,5))
    plt.scatter(
        df_pca['1'],
        df_pca['2'],
        c=df.Group
    )
    plt.show()
    plt.close()
```

## Build Model 2017

```python
df_group = data_original[data_original.Nam == 2017].groupby(['Name', 'Ma_Tinh']).mean().reset_index()
data_2017 = df_group[['So_Diem_Chay','Sunshine','Air_Humidity','Air_Temperature','Rainfull']]
data_2017.head()
```

```python
checkComponentWSSE(data_2017)
```

Từ những biểu đồ trên ta có thể thấy : 
<blockquote>
    - Model K-Mean chạy ỗn định hơn GMM trong trường hợp này. <br>
    - Do năm 2018, 2019 bên dưới chọn phân làm 3 cụm, nên 2017 cũng sẽ chọn như vậy, nhằm giải thích được sự chuyển đổi giữa các cụm qua các năm.
</blockquote>

```python
showTestKMean(data_2017, 10, 'Kmean năm 2017')
```

```python
model_2017 = buildModel(data_2017, 3)
data_2017 = data_2017.assign(
    Ma_Tinh = df_group.Ma_Tinh,
    Ten_Tinh = df_group.Name,
    Group = model_2017.predict(data_2017)
)
data_2017
```

```python
display(
    data_2017[data_2017.Group == 0],
    data_2017[data_2017.Group == 1],
    data_2017[data_2017.Group == 2]
)
```

3 cụm được dược phân ra gồm có:
<blockquote>
    - Cụm 0: Lai Châu, Sơn La với trên 100 diềm cháy mỗi tháng. <br>
    - Cụm 1: Cà Mau, Hà Nội, Nam Định, Tuyên Quang, Vinh, Vũng Tàu, Đà Nẵng với từ 40 điểm cháy trở xuống. <br>
    - Cụm 2: Huế có 48 điểm cháy mỗi tháng.
</blockquote>

```python
showPCA(data_2017)
```

```python
dict_group = {}
for index in range(0, data_2017.shape[0]):
    key = data_2017[['Ma_Tinh', 'Group']].iloc[index, 0]
    value = data_2017[['Ma_Tinh', 'Group']].iloc[index, 1]
    dict_group[key] = value
data_show = data_original[
    data_original.Nam == 2017
].assign(
    Group = data_original.Ma_Tinh.map(lambda x: dict_group[x])
)
# ----------------
data_show.to_csv('data_model/data_corr_2017.csv')
# ----------------
display(
    showCorr('Group 0', data_show[data_show.Group == 0]),
    showCorr('Group 1', data_show[data_show.Group == 1]),
    showCorr('Group 2', data_show[data_show.Group == 2])
)
```

    - Cụm 1: có thể thấy các tỉnh của cụm 1 không liền kề với nhau, bảng hệ số chỉ thấy sự tương quan rõ nét ở một số tỉnh, nhưng đồng thời cũng thấy sự bất thường.

        - Hà nội có số điểm cháy có mối tương quan thuận mạnh với lượng mưa, đây là chổ bất thường.

## Sơn La và Lai Châu

<b> Vì Sơn La và Lai Châu là 2 tỉnh có số điểm cháy nhiều nhất, nên sẽ đặc biệt lấy ra để phân tích riêng. </b>

```python
def funtionExplain(name):
    df = data_original[
        (data_original.Name == name) 
        & (data_original.Nam >= 2017)
    ][[
        'Nam', 'Thang', 'So_Diem_Chay', 'Air_Humidity', 'Sunshine'
    ]]
    df = df.assign(
        Date = df[['Nam', 'Thang']].T.apply(lambda x: date(int(x.Nam), int(x.Thang), 1))
    )
    scale = MinMaxScaler().fit_transform(df[['So_Diem_Chay', 'Air_Humidity', 'Sunshine']])
    df = df.assign(
        So_Diem_Chay = scale[:, 0],
        Air_Humidity = scale[:, 1],
        Sunshine = scale[:, 2]
    )
    plt.figure(figsize=(14, 5))
    plt.plot(
        df.Date,
        df.Sunshine,
        color='y',
        label='Số giờ năng'
    )
    plt.plot(
        df.Date,
        df.So_Diem_Chay,
        color='r',
        label='Số Điểm Cháy'
    )
    plt.plot(
        df.Date,
        df.Air_Humidity,
        color='b',
        label='Độ Ẩm Không Khí'
    )
    plt.title('Số Điểm cháy và Độ ẩm không khí qua các năm của ' + name, fontweight='bold', fontsize=14)
    plt.legend()
    plt.show()
    plt.close()
```

```python
funtionExplain('Sơn La')
```

Nhìn hình trên có thể thấy khi độ ẩm không khí của Son La giảm thấp thì số điểm cháy rừng có xu hướng tăng lên. <br>
Vì số giờ nắng có mối tương quan thuận mạnh vào năm 2017, còn 2018, 2019 thì không, nên có thể xem như 1 yếu tố để tham khảo.
