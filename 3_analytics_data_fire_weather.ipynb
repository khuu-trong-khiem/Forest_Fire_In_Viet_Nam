```python
df_province = pd.read_csv('data_preprocessing/Tinh_Forest.csv', index_col=0)
dict_province = {}
for index in range(df_province.shape[0]):
    row = df_province.iloc[index]
    dict_province[row.values[1]] = row.values[0]
```

```python
def convertProvince(x):
    for k, v in dict_province.items():
        if(x in k):
            return int(v)
    return None

def createCodeProvince(df):    
    df['Code_Province'] = df.Name.map(lambda x: convertProvince(x))
    df = df[~df.Code_Province.isnull()]
    df = df.sort_values(by='Code_Province', ascending=True)
    return df
```

### Understanding Data - General

```python
def filterByName(df, name):
    return df[df.Name == name]

def checkCorr(data1, data2, name):
    df1 = filterByName(data1, name)
    df2 = filterByName(data2, name)
    df_sub = pd.merge(
        df1, 
        df2, 
        left_on=['Year', 'Month'],
        right_on=['Year', 'Month'],
        how='outer',
        suffixes=('_left','_right')
    )
    df_sub = df_sub.dropna()
    return df_sub.value_left.corr(df_sub.value_right)

def getName(df):
    return df[['Name', 'value']].groupby(by=['Name']).count().index.tolist()
```

```python
tuple_name = set(
    getName(data_Average_air_humidity_Year) 
    + getName(data_Average_air_temperature_Year) 
    + getName(data_Mean_sea_level_Year) 
    + getName(data_Total_rainfall_Year) 
    + getName(data_Total_sunshine_hours_Year)  
)
print(tuple_name)
```

```python
list_humidity_temperature = []
list_humidity_sea_level = []
list_humidity_rainfall = []
list_humidity_sunshine = []

list_temperature_sea_level = []
list_temperature_rainfall = []
list_temperature_sunshine = []

list_sea_level_rainfall = []
list_sea_level_sunshine = []

list_rainfall_sunshine = []

for name in tuple_name:
    list_humidity_temperature.append(checkCorr(
        data_Average_air_humidity_Month,
        data_Average_air_temperature_Month,
        name
    ))
    list_humidity_sea_level.append(checkCorr(
        data_Average_air_humidity_Month,
        data_Mean_sea_level_Month,
        name
    ))
    list_humidity_rainfall.append(checkCorr(
        data_Average_air_humidity_Month,
        data_Total_rainfall_Month,
        name
    ))
    list_humidity_sunshine.append(checkCorr(
        data_Average_air_humidity_Month,
        data_Total_sunshine_hours_Month,
        name
    ))
    # -----------
    list_temperature_sea_level.append(checkCorr(
        data_Average_air_temperature_Month,
        data_Mean_sea_level_Month,
        name
    ))
    list_temperature_rainfall.append(checkCorr(
        data_Average_air_temperature_Month,
        data_Total_rainfall_Month,
        name
    ))
    list_temperature_sunshine.append(checkCorr(
        data_Average_air_temperature_Month,
        data_Total_sunshine_hours_Month,
        name
    ))
    # -----------
    list_sea_level_rainfall.append(checkCorr(
        data_Mean_sea_level_Month,
        data_Total_rainfall_Month,
        name
    ))
    list_sea_level_sunshine.append(checkCorr(
        data_Mean_sea_level_Month,
        data_Total_sunshine_hours_Month,
        name
    ))
    # -----------
    list_rainfall_sunshine.append(checkCorr(
        data_Total_rainfall_Month,
        data_Total_sunshine_hours_Month,
        name
    ))
    
df_corr = pd.DataFrame({
    'Name' : list(tuple_name),
    'Humidity vs Temperature' : list_humidity_temperature,
    'Humidity vs Sea level' : list_humidity_sea_level,
    'Humidity vs Rainfall' : list_humidity_rainfall,
    'Humidity vs Sunshine' : list_humidity_sunshine, 
    
    'Temperature vs Sea level' : list_temperature_sea_level,
    'Temperature vs Rainfall' : list_temperature_rainfall,
    'Temperature vs Sunshine' : list_temperature_sunshine,
    
    'Sea level vs Rainfall' : list_sea_level_rainfall,
    'Sea level vs Sunshine' : list_sea_level_sunshine,
    
    'Rainfall vs Sunshine' : list_rainfall_sunshine
})
df_corr = createCodeProvince(df_corr)
df_corr = df_corr[df_corr.columns[0:-1]]
df_corr.head()
```

```python
df_corr_show = df_corr.set_index('Name')
plt.figure(figsize=(9, 7))
sb.heatmap(
    df_corr_show,
    annot=True,
    fmt='0.1f',
    cmap='coolwarm'
)
plt.title('Mối quan hệ giữa các chỉ số thời tiết với nhau qua các tỉnh', fontweight='bold', fontsize=14)
plt.show()
```

Độ ẩm và nhiệt độ có mối tương quan nghịch mạnh ở : Huế, Đà Nẵng, Vinh. <br>
Độ ẩm và lượng mưa có mối thương quan thuận mạnh ở: Cà Mau. <br>
Độ ẩm và số giờ nắng có mối tương quan nghịch ở: Vinh, Huế, Đà Nẵng, Cà Mau. <br>
Nhiệt độ và lượng mưa có mối tương quan thuận mạnh ở : Hà Nội, Tuyên Quang, Nam Định. <br>
Nhiệt độ và số giờ nắng có mối tương quan thuận mạnh ở : Hà Nội, Tuyên Quang, Nam Định, Huế, Đà Nẵng, Vinh.
