```python
import pandas as pd
import glob
import re
```

### Gộp dự liệu theo địa phương

```python
all_files = glob.glob("data_original/toan_quoc/diaphuong*.csv")
all_files[0:5]
```

```python
data_province = pd.read_csv('data_preprocessing/Tinh_Forest.csv', index_col=0)
dict_province = {}
for row in data_province.values:
    dict_province[row[1].lower()] = row[0]
```

```python
data_province.info()
```

```python
data = pd.DataFrame(columns=['Ma_Tinh', 'So_Diem_Chay', 'Nam'])
```

```python
for file in all_files:
    name_file = re.findall('diaphuong_\d+.csv', file)
    year = re.findall('\d+.csv', name_file[0])[0].replace('.csv', '')
    df = pd.read_csv(file, index_col=0)
    df = df.rename(columns={
        'Tên tỉnh' : 'Ten_Tinh',
        'Số điểm cháy': 'So_Diem_Chay'
    })
    df['Nam'] = int(year)
    df['Ma_Tinh'] = df['Ma_Tinh'] = df['Ten_Tinh'].apply(lambda x: dict_province[x.lower()])
    df = df.drop(['#', 'Ten_Tinh'], axis=1)
    data = data.append(df)
```

```python
data = data[data.Ma_Tinh != 0]
data = data.reset_index(drop=True) 
data
```

```python
data.to_csv('data_preprocessing/general_nation_province_year.csv')
```
