```python
import pandas as pd
import numpy as np
```

```python
from lib_analytics import *
```

```python
def convertProvince(x):
    for k, v in dict_province.items():
        if(x in k):
            return int(v)
    return None

def createCodeProvince(df):    
    df['Code_Province'] = df.Name.map(lambda x: convertProvince(x))
    df = df[~df.Code_Province.isnull()]
    df = df.sort_values(by='Code_Province', ascending=True)
    return df
```

### Import Data

```python
df_province = pd.read_csv('data_preprocessing/Tinh_Forest.csv', index_col=0)
dict_province = {}
for index in range(df_province.shape[0]):
    row = df_province.iloc[index]
    dict_province[row.values[1]] = row.values[0]
```

```python
data_fire_province = pd.read_csv('data_preprocessing/general_province_month.csv', index_col=0)
data_air_humidity = pd.read_csv('data_preprocessing/gov/Average air humidity Month.csv')
data_air_temperature = pd.read_csv('data_preprocessing/gov/Average air temperature Month.csv', index_col=0)
data_rainfull = pd.read_csv('data_preprocessing/gov/Total rainfall Month.csv', index_col=0)
data_sunshine = pd.read_csv('data_preprocessing/gov/Total sunshine hours Month.csv')
data_Coordinates = pd.read_csv('data_preprocessing/Coordinates_Provinces.csv', index_col=0)

displaySideBySide({
    'fire forest': data_fire_province.head(3),
    'air humidity': data_air_humidity.head(3),
    'air temperature': data_air_temperature.head(3),
    'rainfull': data_rainfull.head(3),
    'sunshine': data_sunshine.head(3),
    'province': df_province.head(3),
    'coordinates': data_Coordinates.head(3)
})
```

```python
data_air_humidity = createCodeProvince(data_air_humidity)
data_air_temperature = createCodeProvince(data_air_temperature)
data_rainfull = createCodeProvince(data_rainfull)
data_sunshine = createCodeProvince(data_sunshine)
data_Coordinates = createCodeProvince(data_Coordinates)

displaySideBySide({
    'air humidity': data_air_humidity.head(3),
    'air temperature': data_air_temperature.head(3),
    'rainfull': data_rainfull.head(3),
    'sunshine': data_sunshine.head(3),
    'coordinates': data_Coordinates.head(3),
})
```

```python
df_full = pd.merge(
    data_fire_province,
    data_air_humidity,
    left_on=['Nam', 'Thang', 'Ma_Tinh'],
    right_on=['Year', 'Month', 'Code_Province'],
    how='outer',
    suffixes=('','_humidity')
)[[
    'Ma_Tinh', 'Name', 'Nam', 
    'Thang', 'So_Diem_Chay', 'value'
]].dropna()
df_full = df_full.rename(columns={'value': 'Air_Humidity'})
df_full.head(10)
```

```python
df_full = pd.merge(
    df_full,
    data_air_temperature,
    left_on=['Nam', 'Thang', 'Ma_Tinh'],
    right_on=['Year', 'Month', 'Code_Province'],
    how='outer',
    suffixes=('','_Temperature')
)[[
    'Ma_Tinh', 'Name', 'Nam', 
    'Thang', 'So_Diem_Chay', 'value',
    'Air_Humidity'
]].dropna()
df_full = df_full.rename(columns={'value': 'Air_Temperature'})
df_full.head(10)
```

```python
df_full = pd.merge(
    df_full,
    data_rainfull,
    left_on=['Nam', 'Thang', 'Ma_Tinh'],
    right_on=['Year', 'Month', 'Code_Province'],
    how='outer',
    suffixes=('','_Temperature')
)[[
    'Ma_Tinh', 'Name', 'Nam', 
    'Thang', 'So_Diem_Chay', 'value',
    'Air_Humidity', 'Air_Temperature'
]].dropna()
df_full = df_full.rename(columns={'value': 'Rainfull'})
df_full.head(10)
```

```python
df_full = pd.merge(
    df_full,
    data_sunshine,
    left_on=['Nam', 'Thang', 'Ma_Tinh'],
    right_on=['Year', 'Month', 'Code_Province'],
    how='outer',
    suffixes=('','_Temperature')
)[[
    'Ma_Tinh', 'Name', 'Nam', 
    'Thang', 'So_Diem_Chay', 'value',
    'Air_Humidity', 'Air_Temperature',
    'Rainfull'
]].dropna()
df_full = df_full.rename(columns={'value': 'Sunshine'})
df_full.head(10)
```

```python
df_full = pd.merge(
    df_full,
    data_Coordinates,
    left_on=['Ma_Tinh'],
    right_on=['Code_Province'],
    how='outer',
    suffixes=('','_Coordinates')
)[[
    'Ma_Tinh', 'Name', 'Nam', 
    'Thang', 'So_Diem_Chay', 
    'Air_Humidity', 'Air_Temperature',
    'Rainfull', 'Sunshine', 
    'Latitude', 'Longitude'
]].dropna()
df_full.head(10)
```

```python
df_full = df_full.interpolate()
```
