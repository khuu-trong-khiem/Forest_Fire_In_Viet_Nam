```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, date
from dateutil.relativedelta import relativedelta
```

```python
from pmdarima.arima import auto_arima
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import pickle
```

```python
def splitData(df, group):
    df = df_full[df_full.Group == group][['Nam', 'Thang', 'So_Diem_Chay']]
    df = df.assign(
        Date = df[['Nam', 'Thang']].T.apply(lambda x: date(int(x.Nam), int(x.Thang), 1))
    )[[
        'Date', 'So_Diem_Chay'
    ]].set_index('Date')
    return df

def createByYear(model, n):
    list_time = []
    value = model.predict(n_periods=n)
    for x in range(1, n + 1):
        time = df.index[-1] + relativedelta(
            years=1, 
            month=x
        )
        list_time.append(time)
    df_predict = pd.DataFrame({
            'Date': list_time,
            'So_Diem_Chay': value[:n]
    }).set_index('Date')
    return df_predict
```

```python
def createModel(df, seasonal):
    arima = auto_arima(
        df,
        m=12,
        start_p=3,
        start_q=3,
        start_P=0,
        max_p=3,
        max_q=3,
        d=1,
        D=1,
        seasonal=seasonal,
        error_action='ignore',
        suppress_warnings=True,
        stepwise=True
    )
    print('AIC Model :', arima.aic())
    return arima.fit(df)
```

```python
def showFireForest(df, title):
    plt.figure(figsize=(10,5))
    plt.plot(df)
    plt.title(title + ": Fire Forest By Year")
    plt.show()
    plt.close()
    
# Annual maps to 1, quarterly maps to 4, monthly to 12, weekly to 52.
def showSeasonal(df):
    result = seasonal_decompose(x=df, model='multiplicative', period=12)
    plt.rcParams['figure.figsize'] = (10, 7)
    fig = result.plot()
    plt.show()
    plt.close()

def showForcase(df, df_predict, title):
    plt.figure(figsize=(10,5))
    plt.plot(
        df,
        label='Sales',
        color='blue'
    )
    plt.plot(
        df_predict,
        label='Forecase',
        color='red'
    )
    plt.xticks(rotation='vertical')
    plt.title("Fire Forest " + title)
    plt.legend()
    plt.show()
    plt.close()
```

## Time Series - Before 2017

```python
df_full = pd.read_csv('data_model/data_cluster.csv', index_col=0)
```

```python
df_full = df_full[df_full.Nam >= 2017]
```

Trong Trường hợp này sẽ dùng dữ liệu từ năm 2017 trờ đi.

### Time Series for Group 0

```python
df = splitData(df_full, 0)
df = df.groupby(df.index).sum()
df.head(5)
```

```python
showFireForest(df, 'Group 0')
```

```python
showSeasonal(df)
```

Từ biểu đố trên có thể thấy:
<blockquote>
    Treng (xu hướng): bắt đầu tăng từ khoảng giữa năm 2017. <br>
    Seasonal (chu kỳ): bắt đầu thể hiện rỏ ràng từ năm 2017. và không rõ ràng trước năm 2017
</blockquote>

```python
model = createModel(df, True)
# ------
filename = 'save_model/model_timeseries_group_0.sav'
pickle.dump(model, open(filename, 'wb'))
# ------
loaded_model = pickle.load(open(filename, 'rb'))
```

```python
df_predict = pd.concat(
    [
        df.tail(1), 
        createByYear(loaded_model, 3)
    ], 
    axis=0
)
df_predict
```

```python
showForcase(df, df_predict, 'Group 0')
```
